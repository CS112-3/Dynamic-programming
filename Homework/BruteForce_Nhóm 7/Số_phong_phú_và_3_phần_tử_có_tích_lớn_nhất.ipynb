```python
def Help2(n):
    sum=0
    for i in range(1,n):
        if n%i == 0:
            sum=sum+i
            if sum > n:
                return True
    return False

def Help(L,R):
    sum=0
    for i in range(L,R):
        if(Help2(i) == True):
            sum=sum+1
    print(sum)
```
