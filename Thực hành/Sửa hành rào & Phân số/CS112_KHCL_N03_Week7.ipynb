## Đánh giá độ phức tạp thuật toán:
---

* Time Complexity:O(log(n))
* Space Complexity:0(1)

## Abstraction

## Pattern Recognition

Tìm kiếm nhị phân
## Algorithm Designed
---
* Tạo biến Left = giá trị nhỏ nhất giữa 2 mảng 
* Tạo biến Right = giá trị lớn nhất của mảng thứ nhất + giá trị lớn nhất mảng thứ  hai
* Tạo biến Mid = (Left + Right)/2

Tạo một vòng lặp khi Left < Right
* Với từng giá trị Mid, thử giá trị đó làm điều kiện để lấy thanh gỗ 
* Nếu số thanh gỗ được đóng bằng giá trị của những thanh gỗ nhỏ hơn giá trị Mid trong mảng thứ nhất -> chứng tỏ là giá trị tối ưu để làm điều kiện nằm trong khoảng Mid+1 và Right
* Ngược lại thì giá trị tối ưu để làm điều kiện nằm trong khoảng Left và Mid

Sau khi có được độ cao lớn nhất mà hàng rào có thể đạt được = Left -1 
* Dùng giá trị đó để làm điều kiện để lấy thanh gỗ

```markdown
    Lặp từng giá trị của mảng thứ 2 từ index  :

```
# Đánh giá độ phức tạp thuật toán
---
* Time complexity : O(Log(U-V))+ O(NM) 
  * U : giá trị lớn nhất của mảng thứ nhất + giá trị lớn nhất mảng thứ  hai
  * V : giá trị nhỏ nhất trong 2 mảng
  * N,M : số phân tử của từng mảng 
* Space complexity :0(n)

````python

n = int(input())
List1 = [int(i) for i in input().split()]
m = int(input())
List2 = [int(i) for i in input().split()]

counterL2 = 0
Holder=''
left = min(min(List1),min(List2))
right = max(List1)+max(List2)+1
def Test(a ,b , bound):
  temp=[]
  for i in a:
    if i < bound:
      temp.append(i)
	counter=0

	if len(temp) == 0:
		return True

	for j in b:
		if temp[i] + j >= bound:
			counter += 1
		if counter == len(temp):
			return True
	return False
while (left < right):
    mid = (left + right) // 2
    try_ = Test(List1, List2, mid)
    if try_ == True:
        left = mid + 1
    else:
        right = mid
bound = left -1

count=0
for i in range(0, len(List1)):
    if int(List1[i]) >= int(bound):
        continue
    Holder+=str(i+1)+' '
    for j in range(counterL2, len(List2)):
        if int(List1[i]) + int(List2[j]) >= bound:
            counterL2 = j + 1
            Holder+=str(j+1)
            Holder+="\n"
            count+=1
            break
print(bound,count)
print(Holder)
````
